### 1. The Java "ceremony"
Every Java program starts with this structure. For now, accept it as *startup ritual*.
Everything interesting happens inside `main`. 


```
public class Main {
    public static void main(String[] args) {
        System.out.println("Hello EE117");
    }
}
```

Run the cell. You should see output below.


### 2. Variables and Types
Java is **statically typed**. You must declare the type first.


```
public class Main {
    public static void main(String[] args) {
        int voltage = 5;
        double resistance = 220.5;
        String label = "LED";

        System.out.println(label);
        System.out.println(voltage);
        System.out.println(resistance);
    }
}
```

### 3. Simple Calculations
Engineering-style example: Ohm's Law.


```
public class Main {
    public static void main(String[] args) {
        double voltage = 9.0;
        double resistance = 330.0;

        double current = voltage / resistance;

        System.out.println("Current = " + current);
    }
}
```

### Summary
- Java uses a `main` entry point
- Variables have explicit types
- You can compute and print values
