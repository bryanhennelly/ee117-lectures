## Implementing Our New Design - Step 1 â€” The Parent Class

So far, we have only looked at diagrams.

Now we write the code.

We begin with the shared structure - the **parent** class.

---


````java
class LibraryItem {

    public String title;
    public boolean available;

    public void display() {
        System.out.println("Title: " + title);
        System.out.println("Available: " + available);
    }

    public void checkOut() {
        available = false;
    }

    public void returnItem() {
        available = true;
    }
}
````

## Step 2 â€” The Child Classes

Now we define the specialised classes.

They **inherit** from `LibraryItem`.

That means:

- they automatically get `title`
- they automatically get `available`
- they automatically get `display()`
- they automatically get `checkOut()`
- they automatically get `returnItem()`

---

Each child class only needs to add what is unique to itself.

- Book adds `author` and `isbn`
- DVD adds `director` and `releaseYear`
- Magazine adds `issueNumber`

Let us write the three child classes.

---


````java
class Book extends LibraryItem {

    public String author;
    public String isbn;

}

class DVD extends LibraryItem {

    public String director;
    public int releaseYear;

}

class Magazine extends LibraryItem {

    public int issueNumber;

}
````

## Understanding `extends`

In Java, we use:

```java
class Book extends LibraryItem {
}
```

This means:

đź‘‰ `Book` **IS-A** `LibraryItem`

---

## What Happens?

`Book` automatically **inherits everything** from `LibraryItem`:

- data (e.g. `title`, `available`)  
- methods (e.g. `display()`, `checkOut()`, `returnItem()`)  

---

đź‘‰ This is inheritance.

## What This Means

đź‘‰ Data is now **hidden inside the object**

đź‘‰ It can only be accessed through methods

---

This is a big improvement.

---

Butâ€¦

đź‘‰ this means we need a constructor to set the values (and possibly a setter)

But encapsulation with inheritance is a little tricky so we will need to be careful in case we break something.

Lets write the code for the class diagram above with all the prviate attributes and add a constructor...


````java
class LibraryItem {

    private String title; // this time attributes are private
    private boolean available;

    // Constructor
    public LibraryItem(String title) { //so we need a constructor
        this.title = title;
        this.available = true; // default
    }

    public void display() {
        System.out.println("Title: " + title);
        System.out.println("Available: " + available);
    }

    public void checkOut() {
        available = false;
    }

    public void returnItem() {
        available = true;
    }
}
````

````java
// Now lets use inheritance to make the Book
// Since this has new ALSO private attrtibutes we will need a constructor here too...
class Book extends LibraryItem {

    private String author; // our new attributes are ALSO private
    private String isbn;

    public Book(String title, String author, String isbn) { // we need to make a constructor
        this.title = title; // lets try this
        this.author = author;
        this.isbn = isbn;
    }

    public void display() {
        System.out.println("Author: " + title); // lets try this
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}
````

## Problem!! The Parents Private Attributes are blocked even from the child!!!

This means we cannot write this in `Book`'s contructor
```java
this.title = title;
```
nor can we write this in `Book`'s `display()` method

```java
System.out.println("Author: " + title); // lets try this
```

This is because `title` is private to `LibraryItem` and NOT to `Book`.

This can be confusing.... 
- `title` is one of `Book`s attributes but `Book`s methods cannot access it!!!

Dont worry there is a solution
-`super()` to the rescue.

Let's rewrite the two problem lines of code using driect calls to the parent using `super()`

````java
// Now lets use inheritance to make the Book using super()
class Book extends LibraryItem {

    private String author; // our new attributes are ALSO private
    private String isbn;

    public Book(String title, String author, String isbn) { // we need to make a constructor
        super(title); // now lets try this with super()!
        this.author = author;
        this.isbn = isbn;
    }

    public void display() {
        super.display(); // and lets use super here too
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}
````

# 8.4 â€” Constructors and Inheritance

We have just seen something important.

When we wrote this constructor in `Book`:

```java
public Book(String title, String author, String isbn) {
    super(title);
    this.author = author;
    this.isbn = isbn;
}
```

---

## What is `super(title)` doing?

đź‘‰ It calls the **constructor of the parent class**

```java
LibraryItem(String title)
```

---

## Why is this necessary?

Because:

đź‘‰ A `Book` **IS-A** `LibraryItem`

So when we create a `Book`, Java must first:

1. build the **LibraryItem part**
2. then build the **Book part**

---

## Execution Order

When we run:

```java
Book b1 = new Book("The Hobbit", "Tolkien", "123");
```

Java does this:

1. Call `LibraryItem(String title)`  
2. Then run the rest of the `Book` constructor  

---

## Key Idea

đź‘‰ Constructors run **from parent to child**

đź‘‰ The parent must be constructed **first**

---

## Important Rule

If a parent class has a constructor like:

```java
LibraryItem(String title)
```

đź‘‰ Then the child **must call it using `super(...)`**

---

Otherwiseâ€¦

đź‘‰ Java does not know how to construct the parent part  

Lets prove this by breaking the code on purpose...

````java
// Here is the same code as above we have just blocked the super() line in the constructor
// The LibraryItem constructor is not called and...CRASH
class Book extends LibraryItem {

    private String author; // our new attributes are ALSO private
    private String isbn;

    public Book(String title, String author, String isbn) { // we need to make a constructor
        //super(title); // this line is essential
        this.author = author;
        this.isbn = isbn;
    }

    public void display() {
        super.display(); // and lets use super here too
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}
````

## Adding a Default Constructor to the Parent

The code above breaks because when we create a `Book`, Java must:

1. build the **LibraryItem part**
2. then build the **Book part**

---

But it cannot do step 1 because:

đź‘‰ the parent constructor is never called

---

## Why did this suddenly break?

Because earlier we wrote:

```java
public LibraryItem(String title)
```

đź‘‰ This removes Javaâ€™s **default constructor**

---

## Key Idea

đź‘‰ If you define your own constructor  
đź‘‰ Java does NOT provide a default one anymore  

---

So now Java has **no way** to construct `LibraryItem` unless we:

đź‘‰ call `super(...)`  
đź‘‰ OR provide a default constructor ourselves

````java
class LibraryItem {

    private String title; 
    private boolean available;

    // Constructor
    public LibraryItem(String title) { 
        this.title = title;
        this.available = true; // default
    }
    // Lets add a default Constructor
    public LibraryItem() { 
        this.title = null;
        this.available = false; // default
    }

    public void display() {
        System.out.println("Title: " + title);
        System.out.println("Available: " + available);
    }

    public void checkOut() {
        available = false;
    }

    public void returnItem() {
        available = true;
    }
}
````

````java
// Now this code will work - no super() call and still no CRASH
class Book extends LibraryItem {

    private String author; // our new attributes are ALSO private
    private String isbn;

    public Book(String title, String author, String isbn) { 
        //super(title); // this line was essential! but not anymore 
        this.author = author;
        this.isbn = isbn;
    }

    public void display() {
        super.display(); // and lets use super here too
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}

// Prove this works by making a Book object and calling display()



````

A word of caution....The code above works, but it is usually NOT what we want.

We have now lost control of how the parent is initialised.

Using super(...) is still the better design.
