## 5.1 The Problem with Empty Objects

In the last chapter, we created objects like this:

`Book b1 = new Book();`

This creates an object.

But it creates an **empty** object.

It has fields.

But the fields are not properly set yet.

So we do this:

- create the object  
- then fill the fields one by one  

This works.

But it is not the best way. Let's see why.....

---

### Why is this a problem?

Because:

- We can forget to set a field  
- We can create objects that are incomplete  

This is bad design.

We want objects to be correct as soon as they are created.

---

### A small example

We will use a simple Book class.

---

````java
// A simple Book class (no constructor yet)

class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
        System.out.println();
    }
}
````

````java
// Creating an empty object

Book b1 = new Book();

// We must manually fill the fields
// THIS IS A BIT ANNOYING - CONSTURCTORS WILL MAKE THIS LESS TEDIOUS

b1.title = "The Hobbit";
b1.author = "J.R.R. Tolkien";
b1.isbn = "9780261103344";
b1.available = true;

b1.display();
````

````java
// THE REAL PROBLEM: incomplete objects are possible

Book b2 = new Book();

// We forget to set some fields

b2.title = "1984";

// Display anyway
b2.display();
````

Notice what happened.

The program still runs.

But the object is not complete.

This is exactly the problem.

We want to prevent incomplete objects.

## 5.2 What Is a Constructor?

A constructor is a **special method**.

It is used to initialise an object when it is created.

A constructor:

- Has the same name as the class  
- Has no return type (not even `void`)  
- Runs automatically when `new` is used  

You do not call a constructor like a normal method.

It runs when the object is created.

---

Let us see a very small example.

We will use our simple Point class from the last chapter.

---

````java
// A Point class with a constructor

class Point {

    double x;
    double y;

    // Constructor - Note no return type
    Point(double xValue, double yValue) {
        x = xValue;
        y = yValue;
    }

    void display() {
        System.out.println("Point: (" + x + ", " + y + ")");
    }
}
````

````java
// Creating objects

Point p1 = new Point(3.0, 4.0);
Point p2 = new Point(10.0, -2.0);

p1.display();
p2.display();
````

Notice what happened.

When we wrote:
```java
new Point(3.0, 4.0);
```

Java automatically called the constructor.

The values were assigned immediately.

The object was fully initialised at creation.

---

Important:

The constructor has:

- The same name as the class → `Point`
- No return type
- Parameters used to initialise fields

---

Compare this to before:

Previously we did:

- `Point p = new Point();`
- Then set `p.x`
- Then set `p.y`

Now:

The object is complete immediately.

## 5.4 Constructors Enforce Valid Objects

Constructors do more than assign values.

They enforce rules.

They force required data - for example they can stop us setting silly values.

They also prevent incomplete objects.

They make object creation clear and explicit.

An object should begin its life in a valid state.

Not partially built.

---

### Why This Matters

Compare these two styles:

Old style:

- Create empty object  
- Fill fields later  
- Hope nothing is forgotten
- Hope the values make sense

Constructor style:

- Provide required data immediately
- Values MUST make sense  
- Object is valid from the start  

This is stronger design.

---

Let us refactor our Resistor example.

Previously we had:

````java
// OLD VERSION – no constructor

class Resistor {

    double resistance;

    double voltage(double current) {
        return current * resistance;
    }
}

// Usage

Resistor r1 = new Resistor();
r1.resistance = 100;

System.out.println(r1.voltage(2.0));
````

---
This allows:

- Creating a resistor with no resistance set
- Setting resistance later
- Possibly forgetting to set it at all

Now we improve it.

---

### Refactored Version – With Constructor

````java
// NEW VERSION – constructor enforces required data

class Resistor {

    double resistance;

    // Constructor
    Resistor(double r) {
        resistance = r;
    }

    double voltage(double current) {
        return current * resistance;
    }
}
````

---
Notice:

We cannot create a `Resistor` without specifying resistance.

This forces meaningful objects.

The object is valid immediately.

---

### Constructor Improves Clarity

Compare:

`Resistor r = new Resistor(100);`

vs

`Resistor r = new Resistor();`

The first line clearly communicates:

This resistor has 100 ohms.

The second tells us nothing.

Constructors improve readability.

They communicate intent.

And they can alsoe enforce rules abour values!

---

### Refactored Version – Constructor Enforces Rules

Previously, our constructor simply assigned the value.

But a real resistor cannot have negative resistance.

So we improve the constructor.

We enforce a rule:

Resistance must be positive.

---

````java
// Resistor with validation inside constructor

class Resistor {

    double resistance;

    // Constructor with validation
    Resistor(double r) {

        if (r > 0) {
            resistance = r;
        } else {
            System.out.println("Invalid resistance. Must be positive.");
            resistance = 1;   // safe default value
        }
    }

    double voltage(double current) {
        return current * resistance;
    }

    void display() {
        System.out.println("Resistance: " + resistance + " ohms");
    }
}
````

````java
// Test valid resistor

Resistor r1 = new Resistor(100);
r1.display();

System.out.println("Voltage: " + r1.voltage(2.0));
````

````java
// Test invalid resistor

Resistor r2 = new Resistor(-50);
r2.display();

System.out.println("Voltage: " + r2.voltage(2.0));
````

--- 
Now something important has happened.

It is impossible to create a resistor with negative resistance.

The constructor enforces correctness.

The object begins life in a valid state.

This is exactly what constructors are for.

They protect the integrity of the object from the moment it is created.

## 5.5 The Default Constructor

What happens if we do not write a constructor?

Java will create one automatically.

This is called the **default constructor**.

It:

- Takes no parameters  
- Has no body  
- Simply allows object creation
- This is what has been happening in all of our examples before we make contructors - Book etc  

But this only happens under one condition:

If you do **not** write any constructor yourself.

---

### Example 1 – No Constructor Written

````java
// No constructor defined

class Person {

    String name;
    int age;

}

Person p1 = new Person();   // This works

p1.name = "Alice";
p1.age = 25;

System.out.println(p1.name);
````

---
Even though we did not write a constructor, the program works.

Java provided a default no-argument constructor automatically.

---

### Example 2 – When It Disappears

Now watch carefully.

If we write **any constructor**, the automatic default constructor disappears.


````java
class Person {

    String name;
    int age;

    // Custom constructor
    Person(String n) {
        name = n;
    }
}

// This one works
Person p1 = new Person("John")

// But now this now causes an error:
Person p2 = new Person();   // ERROR
````

Why?

Because once we define a constructor ourselves:

Java no longer provides the automatic default constructor.

If we want a no-argument constructor as well, we must write it manually.

---

### Example 3 – Writing Both

````java
class Person {

    String name;
    int age;

    // No-argument constructor
    Person() {
        name = "Unknown";
        age = 0;
    }

    // Parameterised constructor
    Person(String n, int a) {
        name = n;
        age = a;
    }
}

Person p3 = new Person();
Person p4 = new Person("Bob", 30);

System.out.println(p3.name);
System.out.println(p4.name);
````

Summary:

- If you write no constructor → Java creates one automatically.
- If you write any constructor → Java stops creating the default one.
- If you want multiple constructors → you must define them yourself.

Constructors give you control over how objects are created.

And once you take control, Java steps back.

## 5.6 Constructor Parameters

Constructors can take multiple parameters.

Each parameter:

- Receives a value
- Matches to a field
- Initialises the object

The order matters.

Java matches arguments by position.

Not by name.

---

### Book Example with Four Parameters

We now require:

- title  
- author  
- isbn  
- available  

The constructor must receive all four.

---

````java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    // Constructor with four parameters
    Book(String t, String a, String i, boolean avail) {
        title = t;
        author = a;
        isbn = i;
        available = avail;
    }

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
        System.out.println();
    }
}
````

````java
// Creating books with four arguments

Book b1 = new Book(
    "The Hobbit",
    "J.R.R. Tolkien",
    "9780261103344",
    true
);

Book b2 = new Book(
    "1984",
    "George Orwell",
    "9780451524935",
    false
);

b1.display();
b2.display();
````

---

Notice:

The values must be passed in the correct order:

```java
(String, String, String, boolean)
```

If we swap the order incorrectly,
the program will not compile.

Example of incorrect usage - This would cause a compile error

```java
Book b3 = new Book(true, "Title", "Author", "ISBN");
```

Java does not guess.

It matches arguments strictly by position and type.

## 5.7 The `this` Keyword

So far we have written constructors like this:

```java
Book(String t, String a) {
    title = t;
    author = a;
}
```

This works.

But the parameter names (`t`, `a`) are different from the field names (`title`, `author`).

In real programs, we usually often the exact same names for the parameters we pass in as the attribute names.

That creates a problem.

---

### The Problem

We rewrite the example above with the more typical names:

```java
Book(String title, String author) {
    title = title;
    author = author;
}
```

Oh oh !!! There are two different `title` names
- The one inside the class i.e. the data attribute
- The one being passed in as input parameter!!

Which `title` is which?????

The parameter hides the class data attribute - also called `title`.

Inside the constructor, `title` refers to the parameter.

This is where `this` is needed...

---

## What Is `this`?

`this` refers to the current object.

It means:

> The field that belongs to this object.

---

### Correct Version Using `this`

````java
class Book {

    String title;
    String author;

    // Constructor using same parameter names as fields
    Book(String title, String author) {
        this.title = title;
        this.author = author;
    }

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println();
    }
}

Book b1 = new Book("The Hobbit", "J.R.R. Tolkien");
b1.display();
````

---
Now it is clear:

- `this.title` → the data atrribute
- `title` → the parameter being passed in to the contructor

`this` removes ambiguity.

## 5.8 Constructor Overloading

A class can have more than one constructor.

This is called **constructor overloading**.

Overloading means:

- Same constructor name  
- Different parameter lists  

This allows flexible object creation.

We can create objects with different levels of detail.

---

### Example – Book with Two Constructors

Sometimes we may only know:

- title  
- author  

Other times we may also know:

- isbn  

We can support both.

---

````java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    // Constructor 1 – title and author only
    Book(String title, String author) {
        this.title = title;
        this.author = author;
        this.available = true;
    }

    // Constructor 2 – title, author and isbn
    Book(String title, String author, String isbn) {
        this.title = title;
        this.author = author;
        this.isbn = isbn;
        this.available = true;
    }

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
        System.out.println();
    }
}
````

````java
// Using different constructors

Book b1 = new Book("The Hobbit", "J.R.R. Tolkien");

Book b2 = new Book(
    "1984",
    "George Orwell",
    "9780451524935"
);

b1.display();
b2.display();
````

---
Notice:

Java decides which constructor to use based on:

- Number of arguments  
- Types of arguments  

This gives us flexibility.

We can create simple objects.

Or more detailed objects.

## 5.9 Calling One Constructor from Another

When we overload constructors,
we often repeat code.

For example:

- Setting default values  
- Assigning common fields  

This creates duplication.

Java allows one constructor to call another using:

`this(...)`

This must be the **first line** inside the constructor.

---

### Example – Book with Constructor Chaining

````java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    // Main constructor
    Book(String title, String author, String isbn) {
        this.title = title;
        this.author = author;
        this.isbn = isbn;
        this.available = true;
    }

    // Constructor with two parameters - it just call the other constructor above using 'this'
    Book(String title, String author) {
        this(title, author, "Not Assigned");
    }

    // Default constructor
    Book() {
        this("Unknown", "Unknown", "Not Assigned");
    }

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
        System.out.println();
    }
}
````

````java
// Testing constructor chaining

Book b1 = new Book("The Hobbit", "J.R.R. Tolkien", "9780261103344");
Book b2 = new Book("1984", "George Orwell");
Book b3 = new Book();

b1.display();
b2.display();
b3.display();
````

---
Notice:

- The smaller constructors reuse the main constructor.
- We avoid repeating assignment logic.
- All initialisation flows through one place.

Important:

`this(...)` must be the first statement in the constructor.

Constructor chaining like this keeps code nice and clean.

## 5.10 Constructors in Arrays of Objects

We have seen this already but it bears repeating 

When we create an array of objects, something important happens.

The array does **not** create the objects.

It only creates space for references.

It is a common mistake to believe that three objects are constructed when we declare the array using this line.

```java
// Create an array for 3 Book references
Book[] library = new Book[3];
```
At this point:

- The array exists.
- It can hold 3 Book references.
- But there are no Book objects yet.

Each element is currently `null`.

Really the line of code abobe is just step 1. This code is actually correct

---

````java
// STEP 1 Create an array for 3 Book references
Book[] library = new Book[3];


// STEP 2 We must construct each object individually
library[0] = new Book("The Hobbit", "J.R.R. Tolkien");
library[1] = new Book("1984", "George Orwell");
library[2] = new Book("Pride and Prejudice", "Jane Austen");
````

---
Now the array contains three real objects.

Each one was created using a constructor.

---

### Display Using a Loop

````java
// Loop through the array and display each book

for (int i = 0; i < library.length; i++) {
    library[i].display();
}
````

---
Important idea:

- The array holds references.
- Each object must still be constructed.
- Constructors run once per object.

---

### Loop Construction Pattern

Instead of writing each line manually,
we can construct objects inside a loop.

---


````java
// Example with Resistor objects

Resistor[] resistors = new Resistor[5];

for (int i = 0; i < resistors.length; i++) {
    resistors[i] = new Resistor((i + 1) * 100);
}

// Test them
for (int i = 0; i < resistors.length; i++) {
    System.out.println("Voltage: " + resistors[i].voltage(2.0));
}
````

---
This pattern is very common:

1. Create array  
2. Loop  
3. Construct each object  

Remember:

An array of objects does not automatically construct the objects.

You must call the constructor for each one.

## 5.11 Design Thinking: What Should Be Required?

A very impirtant point to remember is that 

**Constructors are design decisions**

When you write a constructor, you decide:

- What must be known at creation?
- What can be optional?
- What can have a default value?

---

### Required Fields

Ask yourself:

> Can this object exist meaningfully without this value?

Example:

A Resistor cannot exist without resistance.

```java
Resistor(double resistance)
```

A Book cannot exist without title and author.

Those should be required.

---

### Optional Fields

Some fields can have safe defaults.

Example:

- available → can default to true  
- isbn → may be optional  

Those do not need to be required.

---

### Good vs Bad Design

Bad design - if owner and balance are essential the this:

```java
BankAccount() { }
```

this allows meaningless objects.

## 5.12 Chapter Summary

In this chapter we learned about **constructors**.

A constructor:

- Has the same name as the class  
- Has no return type  
- Runs automatically when `new` is used  
- Initialises the object  

---

### Why We Use Constructors

Constructors:

- Prevent incomplete objects  
- Require essential data  
- Allow validation  
- Improve design  

An object should start in a valid state.

---

### Key Ideas

- Empty objects are poor design  
- Parameterised constructors require data  
- The default constructor is created only if you write none  
- `this` refers to the current object  
- Constructors can be overloaded  
- `this(...)` lets one constructor call another  
- Arrays of objects require each object to be constructed  

---

### Core Rule

If an object cannot logically exist without a value,

that value belongs in the constructor.

---

Constructors are not just syntax.

They are part of good object design.
