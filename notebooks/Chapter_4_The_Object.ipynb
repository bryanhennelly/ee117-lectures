## 4.2 What Is a Class?

A **class** is a blueprint.

It describes:

- What attributes something has (the data) 
- What it can do (the methods)  

A class does not store real data.

It defines the structure from which objects are created.

---
### 4.2.1 A Very Simple Class

Let us begin with something small.

Suppose we want to represent a point in 2D space.

A point has:

- an x coordinate  
- a y coordinate  

It might also be able to:

- display itself  
- calculate its distance from the origin  

Let us write a class to describe this.

```java
// very simple class
class Point {
    double x;
    double y;
}
```

This defines a **class**.

It is only a blueprint.

Nothing exists in memory yet.

### 4.2.2 Creating an Object

```java
// lets make objects
class Point {
    double x;
    double y;
}

// lets make two objects
Point p1 = new Point();
Point p2 = new Point();

p1.x = 3.0;
p1.y = 4.0;

//Exercise - set values for p2

System.out.println(p1.x);
System.out.println(p1.y);

//Exercise - print valyes for p2
```

---
Now something real exists in memory.

p1 and p2 are both **objects**.

Each has:

- Its own x value  
- Its own y value  

And it can call its own methods...but we dont have any methods yet.

Let's fix that!

Lets give our class its own print behaviour

### 4.2.3 Adding Behaviour

```java
class Point {

    double x;
    double y;

    // our first class method - this just prints the data x and y
    void display() {
        System.out.println("Point: (" + x + ", " + y + ")");
    }

}

Point p1 = new Point();
p1.x = 3.0;
p1.y = 4.0;

// lets use our new method
p1.display();
```

Notice something important:

The display() method does not need x and y as parameters.

Why?

Because the object already contains its own data. 

Lets not add a method that runs a calculation on the data and reurns something...

### 4.2.4 A Method That Computes Something

```java
class Point {

    double x;
    double y;

    void display() {
        System.out.println("Point: (" + x + ", " + y + ")");
    }

    // notice this method has no input (it knows x and y) BUT DOES have an output - a double in this case
    double distanceFromOrigin() {
        return Math.sqrt(x * x + y * y);
    }

}

Point p1 = new Point();
p1.x = 3.0;
p1.y = 4.0;

p1.display();

double d = p1.distanceFromOrigin();
System.out.println("Distance from origin: " + d);
```

The method `distanceFromOrigin()`:

- Uses the object's own data
- Returns a value
- Requires no parameters

This is the key idea:

Data and behaviour belong together.

### 4.3.2 Adding Behaviour – display()

That println statement is annoying

Lets next add print behaviour to the class itself!


```java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    // with one little word - display - we can print all the internal data in a book object!!    
    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
    }
}

// Lets make an object and use display!
Book b1 = new Book();

b1.title = "The Hobbit";
b1.author = "J.R.R. Tolkien";
b1.isbn = "9780261103344";
b1.available = true;

b1.display();
```

Notice again:

The method display() does not need parameters.

The object already has its own data.

### 4.3.3 Adding Behaviour – checkOut()

```java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
    }

    void checkOut() {
        available = false;
    }
}
```

Notice this new method does not take in anything and does not return anything (is `void`) 

But it does CHANGE the value of one attribute!!

```java
Book b1 = new Book();

b1.title = "The Hobbit";
b1.author = "J.R.R. Tolkien";
b1.isbn = "9780261103344";
b1.available = true;

b1.display();

// Now lets check out the book and display again 
b1.checkOut()
b1.display();
    
```

### 4.3.4 Adding Behaviour – returnItem()

There is one method left to do in the class diagram above!

```java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
    }

    void checkOut() {
        if (available) {
            available = false;
        } else {
            System.out.println("Book is already checked out.");
        }
    }
    
    void returnItem() {
        available = true;
    }
}
```
