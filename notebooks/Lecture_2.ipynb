### 1. A First Class
A class models a *thing*.


```
class Motor {
    int rpm;

    void spin() {
        System.out.println("Spinning at " + rpm + " rpm");
    }
}

public class Main {
    public static void main(String[] args) {
        Motor m = new Motor();
        m.rpm = 1200;
        m.spin();
    }
}
```

### Key ideas
- `Motor` is a blueprint
- `m` is an object
- Objects contain data **and** behaviour


### 2. Another Example: LED


```
class LED {
    boolean on;

    void toggle() {
        on = !on;
        System.out.println("LED is " + on);
    }
}

public class Main {
    public static void main(String[] args) {
        LED led = new LED();
        led.toggle();
        led.toggle();
    }
}
```
