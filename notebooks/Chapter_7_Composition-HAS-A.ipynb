# 7.1 From One Object to Many

So far, we have mostly worked with **one object at a time**.

Now we move to something new.

---

We move from:

**One object → Many objects**

Let's define again our old book class and then try to make a collection:

````java
//Simple book class
class Book {
    private String title;
    private String author;
    private boolean available;

    public Book(String title, String author) {
        this.title = title;
        this.author = author;
        this.available = true;
    }

    public void printDetails() {
        System.out.println(title + " by " + author + " is available: " + available);
    }
}

// Declare an array of three book objects
Book[] books = new Book[3]; //NOTE THIS DOES NOT MAKE THREE OBJECTS!!!!! see below
````

At first glance, this might look like we created 3 Book objects.

But this is NOT true.

---

## ⚠️ Important

This line:
```java
Book[] books = new Book[3];
```
does NOT create 3 books.

It creates an array that can hold 3 Book references

---

## Key Idea

We have space for objects  
But no actual objects yet.
So lets make the objects and print their details:

````java
books[0] = new Book("Java", "Smith");
books[1] = new Book("Python", "Lee");
books[2] = new Book("C++", "Khan");

for (int i = 0; i < books.length; i++) {
    books[i].printDetails();
}
````

## What has changed?

Previously:

- We had individual Book objects

Now:

- We have a collection of Book objects

We are starting to build systems, not just single objects.

---


---

#  Building Our First System — Library

So far, we have stored Book objects in an array.

But something is missing.

---

Right now:

- the array exists on its own  
- it is not part of any object  

---

In a real system:

- a Library should **own** the books  

---

## Key Idea

We move from:

array of objects → object that contains objects

````java
class Library {
    private Book[] books;

    public Library() {
        books = new Book[5]; // library can hold 5 books
    }

    public void addBook(Book book, int index) {
        books[index] = book;
    }

    public void printBooks() {
        for (int i = 0; i < books.length; i++) {
            if (books[i] != null) {
                books[i].printDetails();
            }
        }
    }
}
````

Now we have something new.

- A Library object
- That contains Book objects

---

## Key Idea

A Library HAS-A collection of Books.

Lets make a libray objects and put some books inside!

````java
Library lib = new Library();

lib.addBook(new Book("Java", "Smith"), 0);
lib.addBook(new Book("Python", "Lee"), 1);
lib.addBook(new Book("C++", "Khan"), 2);

lib.printBooks();
````

---

## What just changed?

Before:

- We had an array of books

Now:

- The array lives **inside the Library**

---

This is a big step.

---

## Key Idea

The Library is responsible for managing Books.

The books are no longer floating around on their own.

They belong to something.

## Visualising the Relationship

Now we can make the relationship clearer.

A Library does not just have one Book.

It has many Books.

# 7.2 Improving Our Design — From Arrays to ArrayList

## Limitations of Arrays

Our Library works.

But it is not ideal.

---

Arrays allow us to store multiple objects.

But they have problems.

---

### ❌ Fixed size

Once created:

- the size cannot change

---

### ❌ Not flexible

We must:

- manage indexes manually  
- keep track of positions  

---

## Key Idea

Arrays work...

But they are not ideal for building real systems.

---

We need a better way to store collections of objects.

## Introducing ArrayList

Java provides a better tool for storing collections of objects that is very similar to the Python list we learned last semester.

It is called:

**ArrayList**

---

## Key Idea

ArrayList is like an array...

But it can grow and shrink automatically just like a Python list.

Here is how we make one:

````java
import java.util.ArrayList;
ArrayList<Book> books = new ArrayList<>();
````

This creates a new, empty collection of Book objects.

---

## Adding elements

We do not use indexes directly.

Instead we use methods like we did for Python - recall ammend in Python...differnt names for  java

````java
books.add(new Book("Java", "Smith"));
books.add(new Book("Python", "Lee"));
books.add(new Book("C++", "Khan"));

for (int i = 0; i < books.size(); i++) {
    books.get(i).printDetails();
}
````

## Updating Our Library

We are not changing the idea.

A Library still **HAS-A** collection of Books.

---

We are only improving how we store them.

````java
// updated library
import java.util.ArrayList;

class Library {
    private ArrayList<Book> books;

    public Library() {
        books = new ArrayList<>();
    }

    public void addBook(Book book) {
        books.add(book);
    }

    public void printBooks() {
        for (int i = 0; i < books.size(); i++) {
            books.get(i).printDetails();
        }
    }
}
````

````java
// lets add three books
Library lib = new Library();

lib.addBook(new Book("Java", "Smith"));
lib.addBook(new Book("Python", "Lee"));
lib.addBook(new Book("C++", "Khan"));

lib.printBooks();
````

## The relationship has not changed

Even though we changed the implementation:

- Array → ArrayList

The relationship is still:

# 7.3 Multiple Collections: Expanding the System — Members

A given class can conain multiple collections! Lets use our library to explain.

So far, our Library manages Books.

But a real library has more than just books.

It also has people.

---

## Let’s introduce a new class

A Member

A Member represents a person who can borrow books.

````java
import java.util.ArrayList;

class Member {
    private String name;
    private ArrayList<Book> borrowedBooks;

    public Member(String name) {
        this.name = name;
        this.borrowedBooks = new ArrayList<>();
    }

    public void borrowBook(Book book) {
        borrowedBooks.add(book);
    }

    public void printBorrowedBooks() {
        System.out.println(name + " has borrowed:");

        for (int i = 0; i < borrowedBooks.size(); i++) {
            borrowedBooks.get(i).printDetails();
        }
    }
}
````

Now we have something new.

A Member does not just exist.

A Member HAS-A collection of Books.

---

## Key Idea

Objects can contain other objects  
which themselves contain other objects.
Lets visualise this with our class diagram!

```mermaid
classDiagram
Library "1" --> "*" Book : contains
Library "1" --> "*" Member : has
Member "1" --> "*" Book : borrows
```

## This is your most important diagram so far

It shows:

- Library → Books  
- Library → Members  
- Member → Books  

👉 This is a **real system**

---

## Update Library

Now lets rewrite our library code to add members

````java
import java.util.ArrayList;

class Library {
    private ArrayList<Book> books;
    private ArrayList<Member> members;

    public Library() {
        books = new ArrayList<>();
        members = new ArrayList<>();
    }

    public void addBook(Book book) {
        books.add(book);
    }

    public void addMember(Member member) {
        members.add(member);
    }

    public void printBooks() {
        for (int i = 0; i < books.size(); i++) {
            books.get(i).printDetails();
        }
    }
}
````

````java
Library lib = new Library();

Book b1 = new Book("Java", "Smith");
Book b2 = new Book("Python", "Lee");

lib.addBook(b1);
lib.addBook(b2);

Member m1 = new Member("Alice");

lib.addMember(m1);

m1.borrowBook(b1);

m1.printBorrowedBooks();
````

## What just happened?

We now have:

- A Library that manages Books  
- A Library that manages Members  
- Members that manage their own Books  

---

## Key Idea

We are no longer building single classes.

We are building systems of interacting objects.
