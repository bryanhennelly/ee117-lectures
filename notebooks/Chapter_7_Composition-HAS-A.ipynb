So far, we have stored Book objects in an array.

---

In a real system:

- a Library should **own** the books  

---

## Key Idea

We move from:

array of objects → object that contains objects. 

Here is the class diagram again followeed by the code

```mermaid
classDiagram
class Library {
    -books : Book[]
    +addBook(book, index) void
    +printBooks() void
}
```

````java
class Library {
    private Book[] books;

    public Library() {
        books = new Book[5]; // library can hold 5 books
    }

    public void addBook(Book book, int index) {
        books[index] = book;
    }

    public void printBooks() {
        for (int i = 0; i < books.length; i++) {
            if (books[i] != null) {
                books[i].printDetails();
            }
        }
    }
}
````

Now we have something new.

- A Library object
- That contains Book objects

---

## Key Idea

A Library HAS-A collection of Books.

Lets make a libray objects and put some books inside!

````java
Library lib = new Library();

lib.addBook(new Book("Java", "Smith"), 0);
lib.addBook(new Book("Python", "Lee"), 1);
lib.addBook(new Book("C++", "Khan"), 2);

lib.printBooks();
````

---

## What just changed?

- The array lives **inside the Library**

This is a big step.

# 7.2 Improving Our Design — From Arrays to ArrayList

## Limitations of Arrays

Our Library works.

But it is not ideal.

---

Arrays allow us to store multiple objects.

But they have problems.

---

### ❌ Fixed size

Once created:

- the size cannot change

---

### ❌ Not flexible

We must:

- manage indexes manually  
- keep track of positions  

---

## Key Idea

Arrays work...

But they are not ideal for building real systems.

---

We need a better way to store collections of objects.

## Introducing ArrayList

Java provides a better tool for storing collections of objects that is very similar to the Python list we learned last semester.

It is called:

**ArrayList**

---

## Key Idea

ArrayList is like an array...

But it can grow and shrink automatically just like a Python list.

Here is how we make one:

````java
import java.util.ArrayList;
ArrayList<Book> books = new ArrayList<>();
````

This creates a new, empty collection of Book objects.

---

## Adding elements

We do not use indexes directly.

Instead we use methods like we did for Python - recall ammend in Python...differnt names for  java

````java
books.add(new Book("Java", "Smith"));
books.add(new Book("Python", "Lee"));
books.add(new Book("C++", "Khan"));

for (int i = 0; i < books.size(); i++) {
    books.get(i).printDetails();
}
````

## Updating Our Library

We are not changing the idea.

A Library still **HAS-A** collection of Books.

We are only improving how we store them.

Here is the new class diagram followed by the code.

Notice how we write the ArrayList in the Library class.

```mermaid
classDiagram
class Book {
    -title : String
    -author : String
    -available : boolean
    +printDetails() void
}

class Library {
    -books : ArrayList<Book>
    +addBook(book, index) void
    +printBooks() void
}

Library "1" --> "*" Book : contains
```

````java
// updated library
import java.util.ArrayList;

class Library {
    private ArrayList<Book> books; // no more array! 
    // Declare a variable that can refer to a list of Book objects
    // (No list has been created yet!)

    public Library() {
        books = new ArrayList<>(); 
        // Now we CREATE the ArrayList object
        // Only now in the constructor - ie whenever a library object is made!
        // The library now has an empty list ready to store books
    }

    public void addBook(Book book) {
        books.add(book);
    }

    public void printBooks() {
        for (int i = 0; i < books.size(); i++) {
            books.get(i).printDetails();
        }
    }
}
````

````java
// lets add three books
Library lib = new Library();

lib.addBook(new Book("Java", "Smith"));
lib.addBook(new Book("Python", "Lee"));
lib.addBook(new Book("C++", "Khan"));

lib.printBooks();
````

## The relationship has not changed

Even though we changed the implementation:

- Array → ArrayList

The basic relationship is still:

---
Next lets make the libray class - here is the class diagram again:

```mermaid
classDiagram
```

---

````java
import java.util.ArrayList;

class Basket {
    private ArrayList<Item> items;

    public Basket() {
        items = new ArrayList<>();
    }

    public void addItem(Item item) {
        items.add(item);
    }

    public void printItems() {
        for (int i = 0; i < items.size(); i++) {
            items.get(i).printDetails();
        }
    }

    public double getTotalCost() {
        double total = 0;

        for (int i = 0; i < items.size(); i++) {
            total += items.get(i).getPrice();
        }

        return total;
    }
}
````

## 7.6.1 Who Should Do the Work?

Now that our system is more complex, a very important question appears:

👉 **Who should be responsible for borrowing a book?**

---

We have three possible candidates:

- The **Book**
- The **Member**
- The **Library**

---

## Think carefully…

Should we allow this?

```java
member.addBorrowedBook(book);
```

👉 It *looks* reasonable…  
But it causes problems.

---

## The Problem

If a Member borrows a book directly:

- The Book’s availability might not be updated ❌  
- The Library has no control ❌  
- The system can become inconsistent ❌  

---

## The Better Design

👉 The Library should control borrowing

```java
library.borrow(member, book);
```

---

## Why?

Because the Library:

- knows all books  
- knows all members  
- manages the system by
    - checking availability  
    - updating availability  
    - coordinating between objects

---

## Key Idea

👉 The class that manages the system should control interactions  

👉 Other classes should only manage their own data  

---

This is the beginning of **good software design**.

---

## The Real Rule

👉 No object should reach inside another object and change its data directly

👉 It should ask that object to do it

---

This is what keeps systems:

- safe  
- predictable  
- easy to change

This is true **encapsulation**

---

## What does this mean in practice?

Instead of this:

```java
book.available = false; // ❌ direct access
```

We do this:

```java
book.setAvailable(false); // ✅ ask the object to change itself
```

---

👉 We do not touch internal data directly

👉 We use methods to interact with objects


---

## Flow: What happens when a book is borrowed?

Step by step:

1. Library receives request  
2. Library checks if book is available  
3. Book updates its own availability  
4. Member updates its own list  

---

👉 The Library coordinates  
👉 Each object handles its own responsibility  

---

This is how well-designed systems work.
