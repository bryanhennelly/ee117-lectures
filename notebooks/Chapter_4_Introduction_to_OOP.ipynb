## 4.2 What Is a Class?

A **class** is a blueprint.

It describes:

- What attributes something has (the data) 
- What it can do (the methods)  

A class does not store real data.

It defines the structure from which objects are created.

---
### 4.2.1 A Very Simple Class

Let us begin with something small.

Suppose we want to represent a point in 2D space.

A point has:

- an x coordinate  
- a y coordinate  

It might also be able to:

- display itself  
- calculate its distance from the origin  

Let us write a class to describe this.

```java
// very simple class
class Point {
    double x;
    double y;
}
```

This defines a **class**.

It is only a blueprint.

Nothing exists in memory yet.

### 4.2.2 Creating an Object

```java
// lets make objects
class Point {
    double x;
    double y;
}

// lets make two objects
Point p1 = new Point();
Point p2 = new Point();

p1.x = 3.0;
p1.y = 4.0;

//Exercise - set values for p2

System.out.println(p1.x);
System.out.println(p1.y);

//Exercise - print valyes for p2
```

---
Now something real exists in memory.

p1 and p2 are both **objects**.

Each has:

- Its own x value  
- Its own y value  

And it can call its own methods...but we dont have any methods yet.

Let's fix that!

Lets give our class its own print behaviour

### 4.2.3 Adding Behaviour

```java
class Point {

    double x;
    double y;

    // our first class method - this just prints the data x and y
    void display() {
        System.out.println("Point: (" + x + ", " + y + ")");
    }

}

Point p1 = new Point();
p1.x = 3.0;
p1.y = 4.0;

// lets use our new method
p1.display();
```

Notice something important:

The display() method does not need x and y as parameters.

Why?

Because the object already contains its own data. 

Lets not add a method that runs a calculation on the data and reurns something...

### 4.2.4 A Method That Computes Something

```java
class Point {

    double x;
    double y;

    void display() {
        System.out.println("Point: (" + x + ", " + y + ")");
    }

    // notice this method has no input (it knows x and y) BUT DOES have an output - a double in this case
    double distanceFromOrigin() {
        return Math.sqrt(x * x + y * y);
    }

}

Point p1 = new Point();
p1.x = 3.0;
p1.y = 4.0;

p1.display();

double d = p1.distanceFromOrigin();
System.out.println("Distance from origin: " + d);
```

The method `distanceFromOrigin()`:

- Uses the object's own data
- Returns a value
- Requires no parameters

This is the key idea:

Data and behaviour belong together.

### 4.3.2 Adding Behaviour – display()

That println statement is annoying

Lets next add print behaviour to the class itself!


```java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    // with one little word - display - we can print all the internal data in a book object!!    
    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
    }
}

// Lets make an object and use display!
Book b1 = new Book();

b1.title = "The Hobbit";
b1.author = "J.R.R. Tolkien";
b1.isbn = "9780261103344";
b1.available = true;

b1.display();
```

Notice again:

The method display() does not need parameters.

The object already has its own data.

### 4.3.3 Adding Behaviour – checkOut()

```java
class Book {

    String title;
    String author;
    String isbn;
    boolean available;

    void display() {
        System.out.println("Title: " + title);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
        System.out.println("Available: " + available);
    }

    void checkOut() {
        available = false;
    }
}
```

Notice this new method does not take in anything and does not return anything (is `void`) 

But it does CHANGE the value of one attribute!!

```java
Book b1 = new Book();

b1.title = "The Hobbit";
b1.author = "J.R.R. Tolkien";
b1.isbn = "9780261103344";
b1.available = true;

b1.display();

// Now lets check out the book and display again 
b1.checkOut();
b1.display();
    
```

### 4.3.5 From One Book to Many Books

So far we have created a single Book object:

`Book b1 = new Book();`

But a real library does not contain one book.

It contains many books.

How do we store multiple Book objects?

We use an array.

An array can store multiple objects of the same type.

In this case, we want an array of Book objects.

Important:

When we write:

`Book[] library = new Book[3];`

We are creating space for three Book references.

But we are NOT yet creating the actual Book objects.

We must create each object separately.

Let us see this step by step.

```java
// Step 1: Create an array that can hold 3 Book objects

Book[] library = new Book[3];

// Step 2: Create the actual Book objects

library[0] = new Book();
library[1] = new Book();
library[2] = new Book();

// Step 3: Assign values to each book

library[0].title = "The Hobbit";
library[0].author = "J.R.R. Tolkien";
library[0].isbn = "9780261103344";
library[0].available = true;

library[1].title = "1984";
library[1].author = "George Orwell";
library[1].isbn = "9780451524935";
library[1].available = true;

library[2].title = "Pride and Prejudice";
library[2].author = "Jane Austen";
library[2].isbn = "9780141439518";
library[2].available = true;

// Display all books

library[0].display();
library[1].display();
library[2].display();
```

### 4.3.6 A Simple Engineering Example

Let us reinforce the idea with something simpler.

Suppose we want to model a very basic electrical component:

A resistor.

A resistor has:

- one piece of data → resistance (in ohms)

It can:

- calculate voltage from current using Ohm’s Law  
- V = I × R

This is much simpler than our Book class.

But the idea is exactly the same.

Data + behaviour together.

Let us design the class.

```java
class Resistor {

    double resistance;

    double voltage(double current) {
        return current * resistance;
    }

}
```

---
Now let us create one resistor object

---

```java
Resistor r1 = new Resistor();

r1.resistance = 100;   // 100 ohms

double v = r1.voltage(2.0);   // 2 amps

System.out.println("Voltage: " + v);
```

---
That was one object.

Now let us create an array of resistors.

Exactly the same idea as with books.

---

```java
// Create an array for 3 resistors

Resistor[] resistors = new Resistor[3];

// Create the objects

resistors[0] = new Resistor();
resistors[1] = new Resistor();
resistors[2] = new Resistor();

// Assign resistance values

resistors[0].resistance = 100;
resistors[1].resistance = 220;
resistors[2].resistance = 330;

// Calculate voltage for each at 1.5 amps

for (int i = 0; i < resistors.length; i++) {
    double voltage = resistors[i].voltage(1.5);
    System.out.println("Resistor " + i + " Voltage: " + voltage);
}
```

---
Notice:

Each object stores its own resistance.

Each object calculates its own voltage.

The array simply holds references to objects.

The behaviour stays inside the class.

This is the core idea of object-oriented thinking.

## 4.4.4 Implementing the Customer Class

So far, our objects exist independently.

But in a real system, objects interact.

A customer borrows a book.

This means:

- A Customer object must be able to call methods on a Book object.

We will keep this simple.

The Customer will store:

- name
- borrowedCount

And it will have methods:

- borrow(Book b)
- returnBook(Book b)
- display()

Notice:

We are passing an object as a parameter.

```java
class Customer {

    String name;
    int borrowedCount;

    void borrow(Book b) {
        if (b.available) {
            b.checkOut();
            borrowedCount++;
        } else {
            System.out.println("Book is not available.");
        }
    }

    void returnBook(Book b) {
        b.returnItem();
        borrowedCount--;
    }

    void display() {
        System.out.println("Customer: " + name);
        System.out.println("Borrowed count: " + borrowedCount);
    }
}
```

---

## 4.4.5 Demonstrating Object Interaction

Let us now create:

- An array of books
- One customer
- Borrow and return items

Watch how the borrowed count changes across as book objects are borrowed and returned.

```java
// Create books

Book[] library = new Book[2];

library[0] = new Book();
library[1] = new Book();

library[0].title = "The Hobbit";
library[0].author = "J.R.R. Tolkien";
library[0].isbn = "111";
library[0].available = true;

library[1].title = "1984";
library[1].author = "George Orwell";
library[1].isbn = "222";
library[1].available = true;

// Create customer

Customer c1 = new Customer();
c1.name = "Alice";
c1.borrowedCount = 0;

// Borrow a book

c1.borrow(library[0]);

library[0].display();
c1.display();

// Return the book

c1.returnBook(library[0]);

library[0].display();
c1.display();
```

---
Notice what just happened.

- The Customer object changed its borrowedCount.
- The Book object changed its available state.
- One object called methods on another object.

This is how real systems behave.

Objects do not exist alone.

They collaborate.
