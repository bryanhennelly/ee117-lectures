## Look Carefully at the three classes - Book, DVD, Magazine

These are not just related.

They are the **same kind of thing**.

Becuase they share many common attributes and behaviours!

---

We can say:

- A Book **IS-A** library item

- A Magazine **IS-A** library item  

---

## Key Idea

This is a different relationship.

Not:

ūüĎČ HAS-A  

But:

ūüĎČ **IS-A**

---

## Why This Matters

We are rewriting the same methods:

- display()  
- checkOut()  
- returnItem()  

In three different classes.

---

This is a problem.

---

## Key Idea

If multiple classes share the same structure and behaviour:

ūüĎČ We should not rewrite it  

ūüĎČ We should reuse it  

---

## What Comes Next

We are going to:

- remove duplication  
- organise our classes  
- introduce a new structure  

---

This structure is called:

ūüĎČ **Inheritance**

---

And it allows us to say:

ūüĎČ One class builds on another - it INHERITS!

# 8.1 There Must Be a Better Way

We have identified the problem.

---

We are repeating:

- the same data  
- the same methods  
- the same logic  

Across multiple classes.

---

This should feel wrong.

---

## Think About the Design

Right now we have:

- Book  
- DVD  
- Magazine  

Each written separately.

---

But conceptually‚Ä¶

They are all the same kind of thing.

---

## A New Idea

What if we could write the shared parts **once**?

And then reuse them?

---

Instead of this:

- Book has title, available, checkOut()  
- DVD has title, available, checkOut()  
- Magazine has title, available, checkOut()  

---

We could do this:

ūüĎČ Create one general class

ūüĎČ And let the others build on it

---

## Introducing a More General Class

Let us imagine a new class:

ūüĎČ `LibraryItem`

---

This class will contain everything that is common:

- title  
- available  
- display()  
- checkOut()  
- returnItem()  

---

Then:

- Book will build on LibraryItem  
- DVD will build on LibraryItem  
- Magazine will build on LibraryItem

## Implementing Our New Design - Step 1 ‚ÄĒ The Parent Class

So far, we have only looked at diagrams.

Now we write the code.

We begin with the shared structure - the **parent** class.

---


````java
class LibraryItem {

    public String title;
    public boolean available;

    public void display() {
        System.out.println("Title: " + title);
        System.out.println("Available: " + available);
    }

    public void checkOut() {
        available = false;
    }

    public void returnItem() {
        available = true;
    }
}
````

## Step 2 ‚ÄĒ The Child Classes

Now we define the specialised classes.

They **inherit** from `LibraryItem`.

That means:

- they automatically get `title`
- they automatically get `available`
- they automatically get `display()`
- they automatically get `checkOut()`
- they automatically get `returnItem()`

---

Each child class only needs to add what is unique to itself.

- Book adds `author` and `isbn`
- DVD adds `director` and `releaseYear`
- Magazine adds `issueNumber`

Let us write the three child classes.

---


````java
class Book extends LibraryItem {

    public String author;
    public String isbn;

}

class DVD extends LibraryItem {

    public String director;
    public int releaseYear;

}

class Magazine extends LibraryItem {

    public int issueNumber;

}
````

## Understanding `extends`

In Java, we use:

```java
class Book extends LibraryItem {
}
```

This means:

ūüĎČ `Book` **IS-A** `LibraryItem`

---

## What Happens?

`Book` automatically **inherits everything** from `LibraryItem`:

- data (e.g. `title`, `available`)  
- methods (e.g. `display()`, `checkOut()`, `returnItem()`)  

---

ūüĎČ This is inheritance.

## A New Idea

This is a new concept:

ūüĎČ **Method Overriding**

# 8.2 Method Overriding

We now extend our class design so we can make new display methods for each child class.

---

## What Changed?

Each child class now has its **own version of `display()`**.

---

## Key Idea

ūüĎČ A child class can **override** a method from its parent  

ūüĎČ It keeps the same method name  

ūüĎČ But provides its own implementation  

---

This allows each object to behave in its own way.

Lets start by rewriting the Book class:

````java
class Book extends LibraryItem { // still extends!

    public String author;
    public String isbn;

    public void display() { //but now we make our own display just for Book
        System.out.println("Title: " + title);
        System.out.println("Available: " + available);
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}
````

Book now has its **own version** of `display()`.

We hate duplication!!

## Next Idea

What if we could:

ūüĎČ reuse the parent method  
ūüĎČ and still add extra behaviour?

---

That is where `super` comes in.

Let's rewrite Book one more time!

````java
class Book extends LibraryItem {// still extends

    public String author;
    public String isbn;

    public void display() { // we have our own display but.....
        super.display();  // we also reuse parent method
        System.out.println("Author: " + author);
        System.out.println("ISBN: " + isbn);
    }
}
````

## Using `super`

Inside Book we now use:

```java
super.display();
```

---

This means:

ūüĎČ call the **parent version** of display()

---

## What Happens Now?

When we call:

```java
b1.display();
```

Execution is:

1. Run LibraryItem.display()  
2. Then run Book‚Äôs extra lines  

---

## Key Idea

ūüĎČ `super` lets us **reuse the parent method**

ūüĎČ while still **adding new behaviour**

# 8.3 Improving Our Design ‚ÄĒ Encapsulation

So far, our design works.

---

But there is a problem.

Right now, all data is **public**.

---

## Visualising Our Current Design


## The Problem

Because everything is public, we can do this:

```java
b1.available = true;
```

Even if the book is already checked out!

---

ūüĎČ Any code can change any value  
ūüĎČ There is no control  
ūüĎČ There is no protection  

---

This is dangerous.

---

## A Better Idea

We want to:

ūüĎČ protect our data  
ūüĎČ control how it is changed  

---

This met the idea already - its called:

ūüĎČ **encapsulation**

---

## The Change

We will now make our data:

ūüĎČ **private instead of public**

---

## Visualising the New Design

---

## What This Means

ūüĎČ Data is now **hidden inside the object**

ūüĎČ It can only be accessed through methods

---

This is a big improvement.
