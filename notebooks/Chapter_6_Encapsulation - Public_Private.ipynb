### Example 4 – Date (Real-World Rules)

Here is an even better example.....

Dates must follow strict real-world rules.

```mermaid
classDiagram
class Date {
    int day
    int month
    int year
    display()
}
```

For example:

- Month must be between 1 and 12  
- Day depends on the month  
- February has 28 or 29 days  
- Leap years matter  

So let’s again try to enforce these rules in the constructor.

This is a much harder problem than the previous examples.

````java
// Class definition WITH validation (simple if statements only)
class Date {
    int day;
    int month;
    int year;

    Date(int day, int month, int year) {

        this.year = year;

        // Validate month
        if (month >= 1 && month <= 12) {
            this.month = month;
        } else {
            System.out.println("Invalid month - setting to 1");
            this.month = 1;
        }

        // Work out max days in the month
        int maxDays;

        if (this.month == 2) {
            // Leap year check
            boolean leap;

            if (year % 4 == 0) {
                if (year % 100 != 0) {
                    leap = true;
                } else {
                    if (year % 400 == 0) {
                        leap = true;
                    } else {
                        leap = false;
                    }
                }
            } else {
                leap = false;
            }

            if (leap) {
                maxDays = 29;
            } else {
                maxDays = 28;
            }

        } else if (this.month == 4 || this.month == 6 || this.month == 9 || this.month == 11) {
            maxDays = 30;
        } else {
            maxDays = 31;
        }

        // Validate day
        if (day >= 1 && day <= maxDays) {
            this.day = day;
        } else {
            System.out.println("Invalid day - setting to 1");
            this.day = 1;
        }
    }

    void display() {
        System.out.println(day + "/" + month + "/" + year);
    }
}
````

Great — now we have strong rules!

We went to a lot of trouble to make sure we had a good strong rule!!

But it can still be broken!

## 6.2 Encapsulation

In the last section, we discovered a serious problem.

👉 Even when we enforce rules in constructors…

👉 Our objects can still be broken afterwards  

---

### The Core Problem

👉 The data inside our objects is exposed  

👉 Anyone can directly change it  

---

We need a way to **protect the internal data** of an object.

This idea is called:

👉 **Encapsulation**

---

### What Is Encapsulation?

Encapsulation means:

👉 **Keeping data safe inside an object**

👉 **Controlling how that data is accessed and changed**

---

Instead of allowing direct access:

```java
acc.balance = -5000;
```

We will:

👉 **hide the data**

👉 **force all access through controlled methods**

---

### Class Diagrams – New Notation

We now extend our class diagrams.

We introduce two important symbols:

- `-` means **private** → only accessible inside the class  
- `+` means **public** → accessible from outside the class  

---

### Example – Book (Encapsulated)

```mermaid
classDiagram
class Book {
    -String title
    -String author
    -boolean available
    +display()
}
```

👉 The data is now **private**

👉 Only the methods are **public**

We will use these **access specifiers** : **public** and **private** explicity

Note you have avoided using thee before now : the default is **public**

````java
This is a big change.

Previously:

👉 Anyone could access and change fields  

Now:

👉 The object controls access to its own data  

---

This is the key idea:

👉 **Objects protect themselves**
````

````java
// Encapsulated Book class
// Note the use of private for all three data attrbutes
class Book {
    private String title;
    private String author;
    private boolean available;

    Book(String title, String author) {
        this.title = title;
        this.author = author;
        this.available = true;
    }

    void display() {
        System.out.println(title + " - " + author + " (" + available + ")");
    }
}
````

Now let’s try the same thing as before.

Can we still break the object?

````java
Book b = new Book("The Hobbit", "Tolkien");

// Try to break it
b.title = null;   // ❌ This will cause an error
````

👉 The program no longer allows this.

👉 The field is **private**

👉 It cannot be accessed from outside the class  

---

This is exactly what we want!

The object is now protected.

---

### Key Idea

We have changed one thing:

👉 We made the data **private**

---

This means:

👉 Code outside the class can no longer directly change it 

👉 NOTE: methods inside the class itself can of course still access it - eg display()

👉 But the classes own methods are the ONLY place the data can be changed or even accessed for viewing !

👉 The object is now protected  

---

### What Just Happened?

Before:

👉 Anyone could break our objects  

Now:

👉 The object protects itself  

---

### Important

Encapsulation is not about hiding things randomly.

👉 It is about **protecting valid data**

👉 It ensures the object always makes sense  

---

## 6.3 Controlled Access – Getters and Setters

In the previous section, we protected our data using **private fields**.

👉 This stopped objects from being broken  

---

But now we have a new problem…

👉 If everything is private…

👉 How do we access or change the data?

---

We solve this using **methods**.

These methods are called:

👉 **Getters** → used to read data  
👉 **Setters** → used to safely update data  

---

This gives us:

👉 protection  
👉 AND controlled access  

---

This is the full power of encapsulation.

---

### Example 1 – Book

We now upgrade our Book class.

```mermaid
classDiagram
class Book {
    -String title
    -String author
    -boolean available
    +getTitle()
    +setTitle()
    +display()
}
```

👉 The data is private  

👉 We now control access using methods  

---

Let’s define the class.

````java
// Book with getter and setter
class Book {
    private String title;
    private String author;
    private boolean available;

    Book(String title, String author) {
        this.title = title;
        this.author = author;
        this.available = true;
    }

    // Getter
    String getTitle() {
        return title;
    }

    // Setter with validation
    void setTitle(String title) {
        if (title != null) {
            this.title = title;
        } else {
            System.out.println("Invalid title");
        }
    }

    void display() {
        System.out.println(title + " - " + author + " (" + available + ")");
    }
}
````

Now we can safely access and modify the data.

````java
Book b = new Book("The Hobbit", "Tolkien");

// Safe access
System.out.println(b.getTitle());

// Safe update
b.setTitle("New Title");

// Try to break it
b.setTitle(null);   // blocked

b.display();
````

## 6.4 Design Decisions – What Should Be Exposed?

We now understand:

👉 Private fields protect data  
👉 Getters and setters allow controlled access  

---

But now we face a **design question**:

👉 Should we always provide getters and setters for everything?

---

The answer is:

👉 ❌ No

---

Encapsulation is not just about hiding data.

👉 It is about **protecting meaning**

---

### Key Idea

When designing a class, we must decide:

👉 What should be **readable**?  
👉 What should be **modifiable**?  
👉 What should **never change**?  

---

Not all data should have setters.

Some data should be:

👉 Read-only  
👉 Controlled  
👉 Completely hidden  

---

This is where design becomes important.

---

### Example 1 – Book

```mermaid
classDiagram
class Book {
    -String title
    -String author
    -boolean available
    +getTitle()
    +display()
}
```

---

Let’s think:

👉 Should we allow the title to change?

👉 Should we allow the author to change?

---

Usually:

👉 ❌ No

A book does not change its identity.

---

So we might:

👉 Allow reading (getter)  
👉 ❌ Not allow changing (no setter)  

````java
// Book with limited access (no setter)
class Book {
    private String title;
    private String author;
    private boolean available;

    Book(String title, String author) {
        this.title = title;
        this.author = author;
        this.available = true;
    }

    String getTitle() {
        return title;
    }

    void display() {
        System.out.println(title + " - " + author + " (" + available + ")");
    }
}
````

👉 This design prevents accidental changes.

👉 The identity of the book is protected.

👉 That would bypass rules

Instead we design **behaviour**:

👉 deposit()

````java
// BankAccount with behaviour instead of setter
class BankAccount {
    private String owner;
    private double balance;

    BankAccount(String owner, double balance) {
        this.owner = owner;
        this.balance = balance;
    }

    double getBalance() {
        return balance;
    }

    void deposit(double amount) {
        if (amount > 0) {
            balance = balance + amount;
        }
    }

    void withdraw(double amount) {
        if (amount > 0 && amount <= balance) {
            balance = balance - amount;
        } else {
            System.out.println("Invalid withdrawal");
        }
    }

    void display() {
        System.out.println(owner + " : " + balance);
    }
}
````

## 6.5 Design Thinking – Guiding Principles

When designing classes, ask yourself:

---

### 1. Can this value change?

👉 Should this data ever change after creation?

- If **no** → ❌ do NOT provide a setter  
- If **yes** → ✅ consider a setter (with validation)

---

### 2. What rules must always be true?

👉 What makes this object valid?

- Grade must be between 0–100  
- Balance must not be negative  
- Dates must be real  

👉 These rules must be enforced **every time data changes**

---

### 3. Can this field break another field?

👉 Does changing one value affect others?

Example:

- Day depends on month  
- Month depends on year (leap year)

👉 If YES:

❌ Do NOT allow separate setters  

✅ Use a **single method** to update them together  

---

### 4. Are we exposing too much?

👉 Are we giving direct control over important data?

Example:

```java
setBalance(double b)
```

👉 This is dangerous  

Better:

```java
deposit()
withdraw()
```

👉 Provide **behaviour**, not raw access  

---

### 5. Does this reflect the real world?

👉 Think about what the object represents

- A book does not change its author  
- A bank account does not allow random balance edits  
- A date must always be valid  

👉 Good design mirrors real-world logic  

---

## 🔑 Core Idea

Encapsulation is not:

👉 “Make everything private and add setters”

---

Encapsulation is:

👉 **Protecting meaning**  
👉 **Controlling change**  
👉 **Designing safe objects**

## 6.6 Summary – Encapsulation Rules

Let’s summarise what we have learned.

---

### Basic Rules

👉 Fields → **private**  
👉 Methods → **public**

---

### Access Patterns

👉 Read data → use **getters**  
👉 Modify data → use **setters (with validation)**  

---

### Better Design

👉 Do NOT always use setters  

Sometimes we use:

- deposit()
- withdraw()
- borrow()

👉 These represent real-world actions  

---

### Common Mistakes

❌ Making fields public  
❌ No validation in setters  
❌ Allowing invalid states  
❌ Adding setters for everything  

---

### Final Idea

Before:

👉 Objects = simple data holders  

Now:

👉 Objects = controlled systems  

👉 They protect themselves  

👉 They enforce rules  

---

👉 This is **encapsulation**
