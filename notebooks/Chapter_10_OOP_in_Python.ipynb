# 0. Java to Python — What Changes?

The design thinking stays the same.

The syntax changes.

---

## Java

```java
class Book {
    private String title;

    public Book(String title) {
        this.title = title;
    }
}
```

---

## Python

```python
class Book:
    def __init__(self, title):
        self.title = title
```

---

## Key Differences

| Java | Python |
|---|---|
| `{ }` blocks | indentation |
| constructor has class name | constructor is always `__init__` |
| `this.title` | `self.title` |
| `private String title` | `self.title` |
| `extends` | parent class goes in brackets |
| declared variable types | usually no declared variable types |
| `ArrayList<LibraryItem>` | just a Python list |

---

## Important Point

Python does not force as much structure as Java.

That makes Python feel lighter.

But the **class diagram still matters** because it shows the design.

Now let us turn that diagram into Python code.

````python
class Book:

    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = True

    def display(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Available:", self.available)

    def check_out(self):
        self.available = False

    def return_item(self):
        self.available = True
````

## The Special Word `self`

In Python, `self` means:

👉 this particular object

So:

```python
self.title = title
```

means:

👉 store the value `title` inside this object

# 3. Step 2 — Adding Inheritance

Now we introduce the parent class.

---

## 3.1 Class Diagram — Inheritance Structure

This diagram is about **IS-A**.

- A `Book` **IS-A** `LibraryItem`
- A `DVD` **IS-A** `LibraryItem`
- A `Magazine` **IS-A** `LibraryItem`

```mermaid
classDiagram

    -title : str
    -available : bool
    +LibraryItem(title)
    +display()
    +check_out()
    +return_item()
}

    -author : str
    -isbn : str
    +Book(title, author, isbn)
    +display()
}

    -director : str
    -release_year : int

    +Magazine(title, issue_number)
    +display()
}

LibraryItem <|-- Book
LibraryItem <|-- DVD
LibraryItem <|-- Magazine
```

---

Now we implement the diagram in Python.

````python
class LibraryItem:

    def __init__(self, title):
        self.title = title
        self.available = True

    def display(self):
        print("Title:", self.title)
        print("Available:", self.available)

    def check_out(self):
        self.available = False

    def return_item(self):
        self.available = True


class Book(LibraryItem):

    def __init__(self, title, author, isbn):
        super().__init__(title)
        self.author = author
        self.isbn = isbn

    def display(self):
        super().display()
        print("Author:", self.author)
        print("ISBN:", self.isbn)


class DVD(LibraryItem):

    def __init__(self, title, director, release_year):
        super().__init__(title)
        self.director = director
        self.release_year = release_year

    def display(self):
        super().display()
        print("Director:", self.director)
        print("Release year:", self.release_year)


class Magazine(LibraryItem):

    def __init__(self, title, issue_number):
        super().__init__(title)
        self.issue_number = issue_number

    def display(self):
        super().display()
        print("Issue number:", self.issue_number)
````

# 4. Step 3 — Polymorphism in Python

Now we reach the key idea.

In Java, polymorphism often looked like this:

```java
LibraryItem item = new Book(...);
```

The reference type was the parent type.

The actual object was the child type.

---

## But Python Is Different

In Python, we usually do **not** declare the variable type.

So we do not write:

```python
LibraryItem item = Book(...)
```

We simply write:

```python
item = Book(...)
```

---

## So Where Is the Polymorphism?

It appears when we put different objects together and call the same method on them.

---

- If the object is a `Book`, Python runs `Book.display()`
- If the object is a `DVD`, Python runs `DVD.display()`
- If the object is a `Magazine`, Python runs `Magazine.display()`

---

## That Is Polymorphism

One line of code:


Many possible behaviours.

# 5. Duck Typing — Python's Simpler Polymorphism

Now we look at something very Python.

In Java, polymorphism usually depends on inheritance or interfaces.

In Python, polymorphism can be even simpler.

---

## The Duck Typing Idea

Python often cares less about what type an object is.

Python cares more about what the object can do.

---

## The phrase

> If it walks like a duck and quacks like a duck, treat it like a duck.

For us:

👉 if an object has a `display()` method, we can call `display()` on it.

It does not always need to inherit from `LibraryItem`.

---

## 5.1 Class Diagram — No Inheritance, Same Method Name

Look carefully.

There are **no inheritance arrows** here.

But each class still has a `display()` method.

```mermaid
classDiagram

class Book {
    -title : str
    -author : str
    +display()
}

class DVD {
    -title : str
    -director : str
    +display()
}

class Poster {
    -title : str
    -artist : str
    +display()
}

class display_items {
    list of objects with display()
}

display_items --> Book
display_items --> DVD
display_items --> Poster
```

---

This is not the same design as inheritance.

But Python can still use it polymorphically because all objects have `display()`.

````python
class Poster:

    def __init__(self, title, artist):
        self.title = title
        self.artist = artist

    def display(self):
        print("Poster:", self.title)
        print("Artist:", self.artist)


display_items = [
    Book("The Hobbit", "J.R.R. Tolkien", "123"),
    DVD("Inception", "Christopher Nolan", 2010),
    Poster("Library Week", "Design Club")
]

for item in display_items:
    item.display()
    print()
````

---

## 5.2 What Just Happened?

`Poster` does **not** inherit from `LibraryItem`.

But it has a method called:

```python
display()
```

So this still works:

```python
for item in display_items:
    item.display()
```

---

## Important Difference from Java

In Java, we often need a shared parent type or interface.

In Python, we often only need a shared behaviour.

---

## Key Idea

Python polymorphism is often simpler because of **duck typing**.

If the object has the method we call, Python is happy.

# 6. Bringing It Together — The Library Class

Now we combine:

- classes
- inheritance
- composition
- polymorphism

---

## Composition Reminder

Composition means:

👉 one object **HAS-A** another object

For example:

- A `Library` **HAS-A** collection of items

---

## Inheritance Reminder

Inheritance means:

👉 one class **IS-A** kind of another class

For example:

- A `Book` **IS-A** `LibraryItem`
- A `DVD` **IS-A** `LibraryItem`
- A `Magazine` **IS-A** `LibraryItem`

---

Now we combine both ideas.

---

---

### HAS-A

- `Library` HAS-A list of `LibraryItem` objects

---

## The Design Improvement

Instead of:

- one list for books
- one list for DVDs
- one list for magazines

We have:

👉 one list called `items`

---

````python
class Library:

    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_all(self):
        for item in self.items:
            item.display()
            print()
````

---

## 6.2 What Should We Notice?

The `Library` class does not ask:

- is this a book?
- is this a DVD?
- is this a magazine?

It simply stores the object.

Then later it calls:

```python
item.display()
```

---

## This Is the Key Design

The `Library` does not need lots of different lists.

It just needs one list.

## What Did Not Change?

We did not change:

- the loop

---

````python
class AudioBook(LibraryItem):

    def __init__(self, title, narrator, length_minutes):
        super().__init__(title)
        self.narrator = narrator
        self.length_minutes = length_minutes

    def display(self):
        super().display()
        print("Narrator:", self.narrator)
        print("Length in minutes:", self.length_minutes)
````

```python
self._title = title
```

---

Now here is the complete Python version.

````python
class LibraryItem:

    def __init__(self, title):
        self.title = title
        self.available = True

    def display(self):
        print("Title:", self.title)
        print("Available:", self.available)

    def check_out(self):
        self.available = False

    def return_item(self):
        self.available = True


class Book(LibraryItem):

    def __init__(self, title, author, isbn):
        super().__init__(title)
        self.author = author
        self.isbn = isbn

    def display(self):
        super().display()
        print("Author:", self.author)
        print("ISBN:", self.isbn)


class DVD(LibraryItem):

    def __init__(self, title, director, release_year):
        super().__init__(title)
        self.director = director
        self.release_year = release_year

    def display(self):
        super().display()
        print("Director:", self.director)
        print("Release year:", self.release_year)


class Magazine(LibraryItem):

    def __init__(self, title, issue_number):
        super().__init__(title)
        self.issue_number = issue_number

    def display(self):
        super().display()
        print("Issue number:", self.issue_number)


class AudioBook(LibraryItem):

    def __init__(self, title, narrator, length_minutes):
        super().__init__(title)
        self.narrator = narrator
        self.length_minutes = length_minutes

    def display(self):
        super().display()
        print("Narrator:", self.narrator)
        print("Length in minutes:", self.length_minutes)


class Library:

    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_all(self):
        for item in self.items:
            item.display()
            print()
````
